```python
import numpy as np
```

```python
print((np.__version__))         # 1.26.4 (verison of numpy)

# Priority of typcasting during converting into array   ------------->    bool < int < float < complex < str < set < dict

list1 = [1, 4, 5, 6]
print(np.array(list1))          # [1, 4, 5, 6]

list2 = [1, 4.0, 5, 6]
print(np.array(list2))          # [1., 4., 5., 6.]

list3 = [1, 4, "hello", 5, 6]
print(np.array(list3))          # ['1' '4' 'hello' '5' '6']

list4 = [1, 4,8, 5+4j, 6]
print(np.array(list4))          # [1.+0.j 4.+0.j 8.+0.j 5.+4.j 6.+0.j]
```

```python
arr1D = np.array([1,5,8,6,9])
print(arr1D)         # [1 5 8 6 9]
print(arr1D.ndim)
print(arr1D.dtype)
print(arr1D.shape)   # (5,)
print(arr1D.reshape)


arr2D = np.array([
                  [1,2,3],
                  [5,8,11]
                ])
print(arr2D)         # [1 5 8 6 9]
print(arr2D.ndim)
print(arr2D.dtype)
print(arr2D.shape)   # (5,)
print(arr2D.reshape)

arr3D = np.array([
                  [[1,2,7],[5,8,3]],
                  [[8,7,5],[6,3,7]]
                 ])
print(arr3D)         # [1 5 8 6 9]
print(arr3D.ndim)
print(arr3D.dtype)
print(arr3D.shape)   # (5,)
print(arr3D.reshape)

np.arange(0,1001,2)
```
